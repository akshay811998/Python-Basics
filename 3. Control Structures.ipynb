```python
#if-else statement for Diabetic and Obesity Patients
sugarLevel=int(input("Enter your fasting blood sugar level: "))
weight=float(input("Enter your weight: "))
height=float(input("Enter your height: "))
bmi=(weight/(height/100)**2)
if (sugarLevel<150):
    print("Your Sugar Level is Normal")
else:
    print("You're a diabetic, kindly consult a doctor for medications")
if (bmi<18.5):
    print(bmi,"Underweight")
elif (bmi<24.9):
    print(bmi,"Healthy Weight")
elif (bmi<29.9):
    print(bmi,"Over Weight")
else:
    print(bmi,"You're Obese, go to gym")
```

```python
#for loop

lists=[18,23,29,30]
for bmi in lists:
    if (bmi<18.5):
        print(bmi,"Underweight")
    elif (bmi<24.9):
        print(bmi,"Healthy Weight")
    elif (bmi<29.9):
        print(bmi,"Over Weight")
    else:
        print(bmi,"You're Obese, go to gym")
```
